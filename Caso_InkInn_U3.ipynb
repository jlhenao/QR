```python
import numpy      as np
import tensorflow as tf

# 1. Datos (ya cargados en Paso 1)
X = np.array(CANT_CURSOS, dtype=np.float32).reshape(-1, 1)  # feature
y = np.array(PORC_ROTAC,  dtype=np.float32).reshape(-1, 1)  # target

capa   = tf.keras.layers.Dense(units=1, input_shape= [1])
modelo = tf.keras.Sequential([capa])

# 2. Compilación del Modelo
modelo.compile(
   optimizer=tf.keras.optimizers.Adam(0.1),  # Corregido el valor del learning rate
   loss='mean_squared_error'
)

# 4. Entrenamiento del Modelo
print("comenzando")
historial = modelo.fit(CANT_CURSOS, PORC_ROTAC, epochs=1000, verbose=False)
print("modelo entrenado!")
```
